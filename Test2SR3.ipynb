```python
#Load dataset
settlement = pd.read_csv('Informal_Settlement_Programme_-1898186875815614491.csv')
```

```python
#Bar Chart of Basic Services by Settlements: Plot the count of settlements with access to each basic service
#(Water Supply, Sanitation, Refuse, Electricity). This can help identify differences in service provision.
#settlements most lacking in basic services
```

```python
settlement.columns
```

```python
x_columns = settlement[['WATER_SUPP', 'SANITATION', 'REFUSE','ELECTRIC_2']]

plt.figure(figsize=(12.8, 7.2))
for service in x_columns:
    plt.bar(settlement[service].head(2),settlement['EST_HHC'].head(2),alpha=0.7)
plt.title("Basic Services by Settlements")
plt.xlabel("Basic Services")
plt.ylabel("Estimated Household Count")
plt.show()
```

```python
settlement.columns
```

```python
#Scatter Plot of Latitude and Longitude: This can help visualize the spatial distribution of informal settlements on a map, 
#allowing you to identify clustering or dispersion patterns.
```

```python
plt.figure(figsize=(12.8, 7.2))
plt.scatter(settlement['LONGITUDE'].head(25),settlement['LATITUDE'].head(25),color='red')
plt.title("Spatial distribution of informal settlements on a map")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
#Histogram of Estimated Household Count: Show the distribution of household counts in the settlements, providing insights into 
#population density and settlement sizes.
```

```python
data = np.random.normal(loc = 527.6, scale = 971.8, size = 365 )
plt.figure(figsize=(12.8, 2.7))
plt.hist(data, bins = 528, alpha = 0.7, color='red')
plt.ylabel("Frequency")
plt.xlabel("Estimated Household Count")
plt.show()
```
